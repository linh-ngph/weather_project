```python
import requests
import pandas as pd
import time
# Define cities and their coordinates
cities = {
    'New Delhi': {'lat': 28.6139, 'lon': 77.2090},
    'Mumbai': {'lat': 19.0760, 'lon': 72.8777},
    'Kolkata': {'lat': 22.5726, 'lon': 88.3639},
    'Chennai': {'lat': 13.0827, 'lon': 80.2707},
    'Bangalore': {'lat': 12.9716, 'lon': 77.5946}
}
# Define the year range based on FAO data (2002-2022)
years = list(range(2002, 2023))
# Initialize a dictionary to store weather data by year
weather_data_by_year = {'Year': [], 'City': [], 'Average_Temperature': [], 'Average_Humidity': []}
# Fetch weather data from NASA POWER for each city and year
for city, coords in cities.items():
    for year in years:
        # Construct the API URL for NASA POWER (daily climatology for temperature and humidity)
        url = f"https://power.larc.nasa.gov/api/temporal/daily/point?parameters=T2M,QV2M&community=AG&longitude={coords['lon']}&latitude={coords['lat']}&start={year}0101&end={year}1231&format=JSON"
        # Fetch the data from NASA POWER API
        response = requests.get(url)
        if response.status_code == 200:
            data = response.json()
            # Extract temperature (T2M) and humidity (QV2M) from the API response
            temp_data = data['properties']['parameter']['T2M'].values()
            humidity_data = data['properties']['parameter']['QV2M'].values()
            # Calculate average temperature and humidity for the year
            avg_temp = sum(temp_data) / len(temp_data)
            avg_humidity = sum(humidity_data) / len(humidity_data)
            # Append the data to the dictionary
            weather_data_by_year['Year'].append(year)
            weather_data_by_year['City'].append(city)
            weather_data_by_year['Average_Temperature'].append(avg_temp)
            weather_data_by_year['Average_Humidity'].append(avg_humidity)
        else:
            print(f"Failed to fetch data for {city} in {year}, status code: {response.status_code}")
# Convert the dictionary into a DataFrame
df_weather_yearly = pd.DataFrame(weather_data_by_year)
# Display the weather data
print(df_weather_yearly.head(100))
```

```python
df_weather_yearly.to_csv('weather_data.csv')
```
